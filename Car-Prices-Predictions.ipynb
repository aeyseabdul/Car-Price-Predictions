```python
import pandas as pd
import numpy as np
# pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
```

```python
df=pd.read_csv('auto.csv',header=None)
df.head()
```

```python
headers=['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','wheel-base','length','width','height','curb-weight','engine-type','num-of-cylinders','engine-size','fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']
df.columns=headers
```

```python
df
```

```python
df.replace('?',np.nan,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['normalized-losses'].replace(np.nan,df['normalized-losses'].astype('float').mean(axis=0),inplace=True)
df['bore'].replace(np.nan,df['bore'].astype('float').mean(axis=0),inplace=True)
df['stroke'].replace(np.nan,df['stroke'].astype('float').mean(axis=0),inplace=True)
df['horsepower'].replace(np.nan,df['horsepower'].astype('float').mean(axis=0),inplace=True)
df['peak-rpm'].replace(np.nan,df['peak-rpm'].astype('float').mean(axis=0),inplace=True)
df['num-of-doors'].replace(np.nan,'four',inplace=True)
```

```python
df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.isnull().sum()
```

```python
df[["bore", "stroke","price","peak-rpm"]] = df[["bore", "stroke","price","peak-rpm"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
```

```python
dummy_variable_1 = pd.get_dummies(new_df["fuel-type"])
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
```

```python
new_df = pd.concat([new_df, dummy_variable_1], axis=1)
new_df.drop("fuel-type", axis = 1, inplace=True)
```

```python
dummy_variable_2 = pd.get_dummies(new_df["aspiration"])
dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo':'aspiration-turbo'}, inplace=True)
new_df = pd.concat([new_df, dummy_variable_2], axis=1)
new_df.drop("aspiration", axis = 1, inplace=True)
```

```python
X=final_df.copy()
y=X.pop('price')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.score(X_test,y_test)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
y_pred = lm.predict(X_test)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))
```
